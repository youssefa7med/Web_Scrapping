```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

input = input("enter the date of day like this MM/DD/YYYY : ")

url = f"https://www.yallakora.com/match-center/?date={input}"

resp = requests.get(url)

src = resp.content

soup = BeautifulSoup(src,"html")

match_center = soup.find('section',{"class":"matchesCenter"})

all_cards = match_center.find_all('div',{'class':"matchCard"})


with open("yalla kora.csv","w",encoding="UTF-8",newline='') as fd:

    heads = ["نوع الدوري","الفريق الاول","الفريق الثاني","النتيجة"]

    writer = csv.DictWriter(fd,heads)
    writer.writeheader()

    for card in all_cards:
        
        dawry_title = card.find('h2').get_text().strip()
        all_teamA = card.find_all("div",{"class":"teamA"})
        all_teamB = card.find_all("div",{"class":"teamB"})
        all_mresult = card.find_all("div",{"class":"MResult"})

        for i in range(len(all_teamA)):
            teamA = all_teamA[i].find("p").get_text()
            teamB = all_teamB[i].find("p").get_text()
            mresult = f"{all_mresult[i].find_all("span",{"class":"score"})[0].get_text()} - {all_mresult[i].find_all("span",{"class":"score"})[1].get_text()}"

            writer.writerow({"نوع الدوري":dawry_title,"الفريق الاول":teamA,"الفريق الثاني":teamB,"النتيجة":mresult})
            
```
